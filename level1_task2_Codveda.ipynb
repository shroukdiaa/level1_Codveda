# **level1 Basic Task 1**

Description: Preprocess a raw dataset to make it ready
for machine learning

Objectives:

* Handle missing data (e.g., filling with mean/median,
dropping).
* Encode categorical variables (e.g., using one-hot
encoding or label encoding).
* Normalize or standardize numerical features.
* Split the dataset into training and testing sets.
Tools: Python, pandas, scikit-learn


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.preprocessing import MinMaxScaler, LabelEncoder
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder
```

1. load data


```python
# ── Column Names
COLUMNS = [
    'CRIM',    # Per capita crime rate by town
    'ZN',      # Proportion of residential land zoned for large lots
    'INDUS',   # Proportion of non-retail business acres per town
    'CHAS',    # Charles River dummy variable (1 if tract bounds river; 0 otherwise)
    'NOX',     # Nitric oxides concentration (parts per 10 million)
    'RM',      # Average number of rooms per dwelling
    'AGE',     # Proportion of owner-occupied units built prior to 1940
    'DIS',     # Weighted distances to five Boston employment centres
    'RAD',     # Index of accessibility to radial highways
    'TAX',     # Full-value property-tax rate per $10,000
    'PTRATIO', # Pupil-teacher ratio by town
    'B',       # 1000(Bk - 0.63)^2 where Bk is proportion of Black residents
    'LSTAT',   # % lower status of the population
    'MEDV',    # Median value of owner-occupied homes in $1000s (TARGET)
]
TARGET = 'MEDV'
FEATURE_COLS = [c for c in COLUMNS if c != TARGET]
```

```python
df=pd.read_csv("/content/4) house Prediction Data Set.csv",
    header=None,
    names=COLUMNS,
    sep=r'\s+',
    engine='python'
    )
df.head()
```

```python
df.info()
```

```python
df.describe()
```

2. Missing Data

```python
df.sum().isnull()
```

3. Encode Categorical

```python
print("\n[Step 2] Categorical Encoding")

cat_cols = df.select_dtypes(include=['object', 'category']).columns.tolist()

if cat_cols:
    df = pd.get_dummies(df, columns=cat_cols, drop_first=True)
    print(f"  One-hot encoded columns: {cat_cols}")
    print(f"  New shape after encoding: {df.shape}")
else:
    print("  ✓ No string/category columns found.")
    print("  CHAS  → binary (0/1), already numeric")
    print("  RAD   → ordinal integer index, already numeric")
```

4. Standardize Numerical Features

```python
print("\n[Step 3] Standardization (StandardScaler)")

X = df[FEATURE_COLS].copy()
y = df[TARGET].copy()

scaler = StandardScaler()
X_scaled = pd.DataFrame(
    scaler.fit_transform(X),
    columns=FEATURE_COLS
)

print(f"  Standardized {len(FEATURE_COLS)} features → mean ≈ 0, std ≈ 1")
print(f"  Sample means: {X_scaled.mean().round(4).to_dict()}")
```

5. Train / Test Split

```python
print("\n[Step 4] Train / Test Split (80% / 20%)")

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y,
    test_size=0.2,
    random_state=42
)

print(f"  Total samples  : {len(df)}")
print(f"  Training set   : {X_train.shape[0]} samples ({X_train.shape[0]/len(df)*100:.1f}%)")
print(f"  Testing set    : {X_test.shape[0]} samples ({X_test.shape[0]/len(df)*100:.1f}%)")
```

6. Save Outputs

```python
train_df = X_train.copy()
train_df[TARGET] = y_train.values
train_df.to_csv('train_set.csv', index=False)

test_df = X_test.copy()
test_df[TARGET] = y_test.values
test_df.to_csv('test_set.csv', index=False)

print("\n✓ Saved: train_set.csv, test_set.csv")
print("Pipeline complete. X_train, X_test, y_train, y_test are ready for modeling.")
```

# Task
Train and evaluate a Linear Regression model using the preprocessed data from "train_set.csv" and "test_set.csv". The task involves interpreting the model's coefficients and intercept, assessing its performance using R-squared and Mean Squared Error, and visualizing the actual versus predicted values. Finally, summarize the model's insights, coefficients, and evaluation metrics.

## Load Processed Data

### Subtask:
Load the preprocessed training and testing datasets (train_set.csv and test_set.csv) that were generated in the previous step.


**Reasoning**:
The subtask requires loading 'train_set.csv' and 'test_set.csv' into pandas DataFrames. I will use `pd.read_csv` for both files and then display the first few rows of each DataFrame to verify the loading process.



```python
train_df = pd.read_csv('train_set.csv')
test_df = pd.read_csv('test_set.csv')

print("Train DataFrame head:")
print(train_df.head())
print("\nTest DataFrame head:")
print(test_df.head())
```

## Define Features and Target

### Subtask:
Separate the features (X_train, X_test) and the target variable (y_train, y_test) from the loaded training and testing dataframes.


**Reasoning**:
To separate features and target variables, I will drop the 'MEDV' column from the training and testing dataframes to create X_train and X_test, respectively. The 'MEDV' column itself will form y_train and y_test. Finally, I will print the shapes to confirm the split.



```python
TARGET = 'MEDV'

X_train = train_df.drop(columns=[TARGET])
y_train = train_df[TARGET]
X_test = test_df.drop(columns=[TARGET])
y_test = test_df[TARGET]

print("Shape of X_train:", X_train.shape)
print("Shape of y_train:", y_train.shape)
print("Shape of X_test:", X_test.shape)
print("Shape of y_test:", y_test.shape)
```

## Train Linear Regression Model

### Subtask:
Initialize and train a scikit-learn Linear Regression model using the training features (X_train) and target (y_train).


**Reasoning**:
The subtask requires initializing and training a Linear Regression model. I will import `LinearRegression`, create an instance, and then fit it using the `X_train` and `y_train` dataframes already available in the kernel.



```python
from sklearn.linear_model import LinearRegression

# Initialize the Linear Regression model
model = LinearRegression()

# Train the model using the training data
model.fit(X_train, y_train)

print("Linear Regression model trained successfully.")
```

## Interpret Model Coefficients

### Subtask:
Extract and display the intercept and coefficients of the trained linear regression model to understand the relationship between features and the target variable.


**Reasoning**:
To interpret the model coefficients, I will access and display the intercept and then create a pandas Series to associate coefficients with their feature names, finally sorting them by absolute value for easier analysis.



```python
print("Model Intercept:", model.intercept_)

coefficients = pd.Series(model.coef_, index=X_train.columns)

print("\nModel Coefficients (sorted by absolute value):")
print(coefficients.abs().sort_values(ascending=False))
```

## Make Predictions

### Subtask:
Use the trained model to make predictions on the test dataset (X_test).


**Reasoning**:
The subtask requires using the trained model to make predictions on the test dataset. I will use the `predict` method of the `model` object with `X_test` as input.



```python
y_pred = model.predict(X_test)

print("Predictions made successfully.")
print("First 5 predictions:", y_pred[:5])
```

## Evaluate Model Performance

### Subtask:
Calculate and print the R-squared score and Mean Squared Error (MSE) to assess the model's performance on the test set.


**Reasoning**:
The subtask requires calculating and printing the R-squared score and Mean Squared Error (MSE) to evaluate the model. I will import the necessary functions from `sklearn.metrics`, calculate the metrics using `y_test` and `y_pred`, and then print them with appropriate formatting.



```python
from sklearn.metrics import r2_score, mean_squared_error

# Calculate R-squared score
r2 = r2_score(y_test, y_pred)

# Calculate Mean Squared Error (MSE)
mse = mean_squared_error(y_test, y_pred)

print(f"R-squared: {r2:.2f}")
print(f"Mean Squared Error (MSE): {mse:.2f}")
```

## Visualize Predictions

### Subtask:
Create a scatter plot comparing the actual target values (y_test) against the predicted values (y_pred) from the model. Include a diagonal line representing perfect predictions for visual comparison. Make sure to add appropriate labels and a legend.


**Reasoning**:
To visualize the model's predictions against the actual values, I will create a scatter plot as instructed. This requires importing matplotlib, plotting the actual vs. predicted values, adding a reference diagonal line, and labeling the axes.



```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))
plt.scatter(y_test, y_pred, alpha=0.6, label='Actual vs. Predicted')

# Add a diagonal line for perfect predictions
# Determine the min and max values for the line
min_val = min(y_test.min(), y_pred.min())
max_val = max(y_test.max(), y_pred.max())
plt.plot([min_val, max_val], [min_val, max_val], color='red', linestyle='--', label='Perfect Prediction Line')

plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')
plt.title('Actual vs. Predicted Values for Linear Regression Model')
plt.legend()
plt.grid(True)
plt.show()
```
